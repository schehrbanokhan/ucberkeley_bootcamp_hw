# WeatherPy

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
saved_responses=[]
counter=1
set_no=1

print("Beginning Data Retrieval")
print("-------------------------------------------------")

for x in range(len(cities)):
    
    para_1={"appid":api_key,
       "q":cities[x]}
    url_1="http://api.openweathermap.org/data/2.5/weather"
    api_1=requests.get(url_1,params=para_1).json()
    
    if counter==51:
        set_no+=1
        counter=1
    
    if api_1["cod"] == int(200):
        print(f"Printing record {counter} of set {set_no} | {cities[x]}")
        
    else:
        print("City not found. Skipping...") 
    
    counter+=1
    saved_responses.append(api_1)
        
print("-------------------------------------------------")
print("Data Retrieval Complete ")
print("-------------------------------------------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#To create a data frame with all the city names but other empty columns
df_city=pd.DataFrame({"City":cities})

df_city["Cloudiness"]=""
df_city["Country"]=""
df_city["Date"]=""
df_city["Humidity"]=""
df_city["Latitude"]=""
df_city["Longitude"]=""
df_city["Max. Temperature"]=""
df_city["Wind Speed"]=""

df_city.head()
```

```python
for index, row in df_city.iterrows():

    try:
        df_city.loc[index, 'Cloudiness'] = saved_responses[index]['clouds']['all']
        df_city.loc[index, 'Country'] = saved_responses[index]['sys']['country']
        df_city.loc[index, 'Date'] = saved_responses[index]['dt']
        df_city.loc[index, 'Humidity'] = saved_responses[index]['main']['humidity']
        df_city.loc[index, 'Latitude'] = saved_responses[index]['coord']['lat']
        df_city.loc[index, 'Longitude'] = saved_responses[index]['coord']['lon']
        df_city.loc[index, 'Max. Temperature'] = saved_responses[index]['main']['temp_max']
        df_city.loc[index, 'Wind Speed'] = saved_responses[index]['wind']['speed']
        
    except (KeyError, IndexError):
        pass
```

### Plotting the Data

```python
#Removed cities from the data frame that donot have any numeric values in them
df_city["Country"].replace('',np.nan,inplace=True)

df_city.dropna(axis=0,how="any",inplace=True)

df_city.head()
```

#### Latitude vs. Temperature Plot

```python
lat=df_city["Latitude"]
max_temp=df_city["Max. Temperature"].tolist()

new_f=[]

for x in range(len(max_temp)):
    y=(1.8*(max_temp[x]-273.15))+32
    new_f.append(y)
new_f

plt.scatter(lat,new_f,marker="o",edgecolors="black",alpha=0.75)
plt.grid()
plt.title("Max Temperature(F) vs. Latitude (01/06/2019)")
plt.ylabel("Max. Temperature (F)")
plt.xlabel("Latitude")
plt.savefig("MaxTemp.png")
```

#### Latitude vs. Humidity Plot

```python
lat=df_city["Latitude"]
hum=df_city["Humidity"]

plt.scatter(lat,hum,marker="o",edgecolors="black",alpha=0.75)
plt.grid()
plt.ylim(-10,110)
plt.title("Humidity vs. Latitude (01/06/2019)")
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")
plt.savefig("Humidity.png")
```

```python
lat=df_city["Latitude"]
wind=df_city["Wind Speed"]

plt.scatter(lat,wind,marker="o",edgecolors="black",alpha=0.75)
plt.grid()
plt.title("Wind Speed vs. Latitude (01/06/2019)")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed (mph)")
plt.savefig("Windspeed.png")
```
